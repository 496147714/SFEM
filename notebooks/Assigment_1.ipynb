### (b) Plot the first five eigenfunctions and the relative variance error with five terms in the expansion.

Define two error functions: `relVarianceErrorKL` and `avgVarianceErrorKL`:

```python
def relVarianceErrorKL(m, L, l, Lambda, phi, z):
    """
    Returns the relative variance error of the KL expansion with
    exponential kernel
    
    Parameters
    ----------
    m: number of random variables
    L: Length of the domain
    l: correlation length
    lambda: vector of eigenvalues
    phi: cell array of eigenfunctions
    z: longitudinal coordinate

    Returns
    -------
    errVar : 2-rank array 
         relative variance error
    """
    
    errVar = np.ones((len(z),m))
    errVar[:,0] = errVar[:,0] - Lambda[0]*phi[0](z)**2
    for i in range(1,m):
        errVar[:,i] = errVar[:,i-1] - Lambda[i]*phi[i](z)**2

    return errVar


def avgVarianceErrorKL(m, L, Lambda):
    """
    Returns the average variance error of the KL expansion with
    exponential kernel

    Parameters
    ----------
    m: number of random variables
    L: Length of the domain
    l: correlation length
    lambda: vector of eigenvalues
    phi: cell array of eigenfunctions

    Returns
    -------
    errVar : 1-rank array 
         average variance error
    """

    errVar = np.ones(m);
    errVar0 = 1.
    errVar[0] = errVar0 - 1./L * Lambda[0]
    for i in range(1,m):
        errVar[i] = errVar[i-1] - 1./L * Lambda[i]

    return errVar
```

### Visualization of the goodness of the KL decomposition for an exponential auto-correlation function

#### Exact exponential auto-correlation function

````python
# Exact auto-correlation function
def auto_correlation(x,y,l_corr,sigma=1.):
    """
    exponential auto-correlation
    """
    distance = np.sqrt((x-y)**2)
    return sigma**2 * np.exp(-distance/l_corr)
````

````python
# make a mesh
x = np.linspace(0.,2.,101)
X, Y = np.meshgrid(x,x)
````

````python
# Evaluate auto-correlation function
rho_XY = auto_correlation(X, Y, l)
````

````python
plt.figure()
plt.imshow(rho_XY, extent=[0.,2.,0.,2.])
````

#### Spectral decomposition of exponential auto-correlation function

````python
# Define the KL approximation of the exponential auto-correlation function
def auto_correlation_KL(x, y, Lambda, phi, m):
    """
    exponential auto-correlation. KL decomposition
    """
    distance = np.linalg.norm(x-y)
    AC = np.zeros(np.hstack((m,X.shape)))
    AC[0] = AC[0] + Lambda[0]*phi[0](x)*phi[0](y)
    for i in range(m):
        AC[i] = AC[i-1] + Lambda[i]*phi[i](x)*phi[i](y)
    return AC
````

````python
# Evaluate auto-correlation function
m = 9
rho_XY_KL = auto_correlation_KL(X, Y, Lambda, phi, m)
print rho_XY_KL.shape
````

````python
# Plot the first "m" approximations
plt.figure(figsize=(12,12))
for i in range(m):
    plt.subplot(3,3,i+1)
    plt.imshow(rho_XY_KL[i], extent=[0.,2.,0.,2.])
    plt.title('KL decomposition, m=%s' % (i+1))
plt.tight_layout()
````

#### Pointwise difference between extact and approximated AC function

````python
# Plot the first "m" approximations
plt.figure(figsize=(12,12))
for i in range(m):
    plt.subplot(3,3,i+1)
    plt.imshow(rho_XY-rho_XY_KL[i], extent=[0.,2.,0.,2.])
    plt.title('KL decomposition, m=%s' % (i+1))
plt.tight_layout()
````
