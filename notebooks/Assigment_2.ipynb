(c) Solve the global system of equations to obtain the random realizations of

### (a) Compute a random realization of the axial resistance.

````python
# import the function EigFcnKL in module EigFcnKL
from EigFcnKL import EigFcnKL
from relVarianceErrorKL import relVarianceErrorKL
from avgVarianceErrorKL import avgVarianceErrorKL

# maximum number of terms of the KL expansion
m_max = 100

# eigenvalues and eigenfunctions of the KL expansion with exponential kernel
Lambda, phi, w, alpha = EigFcnKL(m_max , L , l)

# Make a grid to evaluate the eigenfunctions and the errors
N_nodes    = 21
N_elements = N_nodes - 1
z     = np.linspace(0, L, N_nodes)
z_mid = (z[:-1] + z[1:]) / 2.

# relative variance error at midpoints
errVar = relVarianceErrorKL(m_max, L, l, Lambda, phi, z_mid)

# average variance error at midpoints
avgErr = avgVarianceErrorKL(m_max, L, Lambda)

# D(z,theta) is described by a homogeneous RF with lognormal marginal distribution
# Transform the parameters of the lognormal distribution to a normal distribution
sigma_U =  np.sqrt(np.log(delta_D**2 + 1.))
mu_U =  np.log(mu_D) - sigma_U**2/2.

# find the first m-terms that ensure an average variance error < 0.05 
avgErr_max = 0.05

m = 0
while m < m_max:
    m = m + 1
    if avgErr[m] < avgErr_max:
        break
print 'N_terms, avgErr[m]', m+1, avgErr[m]

# Number of realizations
# N_realizations = 1
N_realizations = 1

# generate the random numbers that are gonna be used in the simulations
np.random.seed(123)
UU = np.random.randn(N_realizations,m)

plt.figure()

for i in range(N_realizations):
    
    U = UU[i,:]

    U_RF = np.zeros(len(z_mid))

    for i in range(m):
        U_RF = U_RF + np.sqrt(Lambda[i])*phi[i](z_mid)*U[i]

    U_RF = U_RF * sigma_U
    U_RF = U_RF + mu_U

    # figure()
    # plot(z, U_RF)

    D_RF = np.exp(U_RF)

    plt.plot(z_mid, D_RF)

plt.plot(z_mid, np.ones_like(z_mid)*mu_D,'k',linewidth=3,label='mean')
plt.legend(loc='best')
plt.xlim(0.,2.)
plt.xlabel('distance [m]', fontsize='large')
plt.ylabel('axial resistance [kN]', fontsize='large')
plt.title('KL expasion: 1 realizations of axial resistance RF', fontsize='large')
````

### (b) Compute the element stiffness matrices and assemble the global matrices. Apply the midpoint method to perform the integration for evaluating the element stiffness matrices.

````python
# import functions febar and assembleK
from febar import febar
from assembleK import assembleK

# element length
h = 1. * L / N_elements

# element stiffness matrix
D = D_RF
Ke = 1.* D / h

# assemble stiffness matrix
K = assembleK(Ke, N_elements, N_nodes)
````

Global stiffness matrix is sparse (see `assembleK` function):

```Python
import numpy as np
from scipy import sparse

def assembleK(Ke, nele, nnode):
    """
    assembles the global stiffness matrix of bar
    Ke: vector of element stiffness
    nele: number of elements
    nnode: number of nodes
    """

    # initialize global stiffness matrix of bar
    K = np.zeros((nnode,nnode))
    K = sparse.csr_matrix(K, shape=(nnode,nnode))

    # assembles the global stiffness matrix
    idx  = range(0,nele)
    idx2 = range(1,nele+1)
    K = K + sparse.csr_matrix(( Ke, (idx, idx )), shape=(nnode,nnode))
    K = K + sparse.csr_matrix((-Ke, (idx, idx2)), shape=(nnode,nnode))
    K = K + sparse.csr_matrix((-Ke, (idx2,idx )), shape=(nnode,nnode))
    K = K + sparse.csr_matrix(( Ke, (idx2,idx2)), shape=(nnode,nnode))

    return K
```

### (d.1) Investigate the influence of the number of elements on the accuracy of the displacement at the tip of the bar.

````python
# Make a grid to evaluate the eigenfunctions and the errors
nElements = np.array([10, 20, 50, 100, 1000])

# D(z,theta) is described by a homogeneous RF with lognormal marginal distribution
# Transform the parameters of the lognormal distribution to a normal distribution
sigma_U =  np.sqrt(np.log(delta_D**2 + 1.))
mu_U =  np.log(mu_D) - sigma_U**2/2.

# correlation length
l = 1.

# random seeds
seeds = [123, 12345, 1234567, 123456789, 384859]

# accuracy threshold
avgErr_max = 0.05

# errors 
errVar = []
avgErr = []

# statistics
mean_U  = np.zeros(len(nElements))
std_U   = np.zeros(len(nElements))
delta_U = np.zeros(len(nElements))

# Number of realizations
N_realizations = 20

# initialize tip displacement
u_tip   = np.zeros((len(nElements), N_realizations))

# define figure
plt.figure(figsize=(16,12))

# loop over correlation legths
for j,N_elements in enumerate(nElements):

    # calculate nodes and midpoints
    N_nodes = N_elements + 1
    z       = np.linspace(0, L, N_nodes)
    z_mid   = (z[:-1] + z[1:]) / 2.
   
    # eigenvalues and eigenfunctions of the KL expansion with exponential kernel
    Lambda, phi, w, alpha = EigFcnKL(m_max , L , l)

    # average variance error at midpoints
    avgErr.append(avgVarianceErrorKL(m_max, L, Lambda))
    
    # find the first m-terms that ensure an average variance error < 0.05 
    m = 0
    while m < m_max:
        m = m + 1
        if avgErr[j][m] < avgErr_max:
            break
    print 'N_elements=%4i, N_terms=%2i, avgErr[m]=%.3f,' % (N_elements, m+1, avgErr[j][m]),

    # generate the random numbers that are gonna be used in the simulations
    # a different sequence for different realizations
    np.random.seed(seeds[j])
    UU = np.random.randn(N_realizations,m)

    # loop over realizations
    for i in range(N_realizations):

        # Gaussian RVs
        U = UU[i,:]

        # Initialize Gaussian RF 
        U_RF = np.zeros(len(z_mid))

        # Calculate Gaussian RF 
        for ii in range(m):
            U_RF = U_RF + np.sqrt(Lambda[ii])*phi[ii](z_mid)*U[ii]
            
        U_RF = U_RF * sigma_U
        U_RF = U_RF + mu_U

        # figure()
        # plot(z, U_RF)

        # Transformation to distribution of axial resistance (lognormal)
        D_RF = np.exp(U_RF)

        # finite element solution (random realization)
        u_RR = febar(D_RF, q, L, N_elements)

        # store tip displacements
        u_tip[j,i] = u_RR[-1]
        
        # plot random realizations of bar displacement
        if j+1<len(nElements):
            plt.subplot(2,2,j+1)
            plt.plot(z, u_RR)
            plt.xlim(0., 2.)
            plt.ylim(0., 0.04)
        
            # plot annotations only once per correlation length
            if j>1:
                plt.xlabel('distance [m]', fontsize='large')
            if np.mod(j,2)==0:
                plt.ylabel('displacement [m]', fontsize='large')
            plt.title('KL decomposition, N_elements=%s' % (N_elements), fontsize='large')
        else:
            # reference
            plt.figure(33)
            plt.subplot(111)
            plt.plot(z, u_RR)
            plt.xlim(0., 2.)
            plt.ylim(0., 0.04)
            plt.xlabel('distance [m]', fontsize='large')
            plt.ylabel('displacement [m]', fontsize='large')
            plt.title('KL decomposition, N_elements=%s' % (N_elements), fontsize='large')
        
    u_tip_mean = np.mean(u_tip[j,:])
    u_tip_std  = np.std(u_tip[j,:])
    u_tip_CV   = u_tip_std / u_tip_mean
    print 'mean (u_tip)=%.6f, CV (u_tip)=%.6f' % (u_tip_mean, u_tip_CV)

````

#### Results:
- Average variance error do not depend on number of elements
- The larger the number of elements, the closer the mean of the random realizations
  to the reference (N_elements=1000, u_tip_mean=0.020142) and to the deterministic
  tip displacement (u_tip=0.02m).
- Coefficient of variation is comparable for all cases. However, it seems that the
  for increasing number of elements, the variance is slighly the smaller (less fluctuations).
- Pseudo-random RN sequences are not the best suited for sampling.
